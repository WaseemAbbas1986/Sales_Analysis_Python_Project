# Exploratory Data Analysis

```python
ax = sns.countplot(data = df, x = 'Gender', hue = 'Gender',palette='Set1')
plt.title(" Order Count By Gender")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
df.columns
```

```python
ax = sns.countplot(x ='Age Group', data = df,hue = 'Gender',palette='deep')
plt.title(" Order Count By Age group & Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_fig = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(['Amount'],ascending=False)
ax = sns.barplot(x= 'Age Group', y = 'Amount', data =sales_fig,hue = 'Age Group', palette='Set1' )
plt.title(" Sales By Age group")
```

```python
Order_State = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(['Orders'],ascending=False).head(10)
sns.set(rc= {'figure.figsize':(15,10)})
ax = sns.barplot(x= 'State', y = 'Orders', data =Order_State,hue = 'State', palette='deep' )
plt.title(" Total Orders By State (Top 10)")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
Order_State = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(['Amount'],ascending=False).head(10)
sns.set(rc= {'figure.figsize':(15,10)})
ax = sns.barplot(x= 'State', y = 'Amount', data =Order_State,hue = 'State', palette='deep' )
plt.title(" Total Sales By State (Top 10)")
plt.show()
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status', hue = 'Marital_Status',palette='Set2')
plt.title(" Order Count By Marital Status")
for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc= {'figure.figsize':(5,8)})
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status', hue =  'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
ax = sns.countplot(data = df, x = 'Occupation', hue =  'Gender',palette= 'deep')
for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc= {'figure.figsize':(20,10)})
plt.title("Order Count By Ocupation & Gender")
plt.show()
```

```python
ax = sns.countplot(data = df, x = 'Product_Category',hue = 'Product_Category',palette='deep')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show(10)
```

```python
Order_State = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(['Amount'],ascending=False).head(10)
sns.set(rc= {'figure.figsize':(20,5)})
ax = sns.barplot(x= 'Product_Category', y = 'Amount', data =Order_State,hue = 'Product_Category', palette='deep' )
plt.title(" Total Sales By Product Catagory (Top 10)")
plt.show()
```
